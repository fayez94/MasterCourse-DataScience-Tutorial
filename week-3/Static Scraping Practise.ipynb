```python
#the code at a time
web_page = bs4.BeautifulSoup(requests.get(url, {}).text, "lxml")
tables = web_page.find_all(name="table", attrs={"class": "wikitable"})
table_rows = tables[-1].find_all(name = "tr")
len(table_rows)
```

```python
col_names = table_rows[0].find_all(name="th")
col_names[0]
```

```python
col_names[0].text
```

```python
col_names[0].text.replace("\n", "")
```

```python
col_names = [raw_name.text.replace("\n", "") for raw_name in col_names]
col_names
```

```python
#together
col_names = [raw_name.text.replace("\n", "") for raw_name in table_rows[0].find_all(name="th")]
col_names
```

```python
contents = table_rows[2].find_all(name="td")
contents[0].text
```

```python
contents[0].text.replace("\n","")
```

```python
contents[0].text.replace("\n","").strip()
```

```python
#for single row
contents = [raw_name.text.replace("\n", "").strip() for raw_name in table_rows[2].find_all(name="td")]
contents
```

```python
#for all rows
contents = []
for row_id in range(2, len(table_rows)):
  content = [row_name.text.replace("\n", "").strip() for row_name in table_rows[row_id].find_all(name="td")]
  contents.append(content)
contents[2]
```

```python
#put the table data in a column wise (for single row)
x_dict = {}
for idx, row_name in enumerate(table_rows[2].find_all(name="td")):
  x_dict[col_names[idx]] = row_name.text.replace("\n", "").strip()
x_dict

```

```python
#put the table data in a column wise (for all row)
contents = []
for row_id in range(2, len(table_rows)):
  content = {col_names[idx]: row_name.text.replace("\n", "").strip() for idx, row_name in enumerate(table_rows[idx].find_all(name="td"))}    #dictionary comprehension
contents.append(content)
```

```python
contents[-1]
```

```python
#previos loop in comprehension
contents = [{col_names[idx] : raw_name.text.replace("\n", "").strip() for idx, raw_name in enumerate(table_rows[row_id].find_all(name="td"))} for row_id in range(2, len(table_rows))]
```

```python
contents[-5]
```

```python
df = pd.DataFrame(data = contents, columns = col_names)
df.head()
```

```python
len(df)
```

```python
df.to_csv("Static_Scraping.csv")
```
