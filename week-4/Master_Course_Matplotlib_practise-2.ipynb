```python
df = pd.read_csv("/content/best_cs_scientific_details.csv")
df
```

## Line Chart

```python
plt.rcParams["figure.figsize"] = plt.rcParamsDefault["figure.figsize"]
```

```python
country_cnt = df['Country'].value_counts().to_dict()

countries = list(country_cnt.keys())
cnt = list(country_cnt.values())

plt.rcParams["figure.figsize"] = (15,5)

plt.title("Presence of Countries")

plt.plot(cnt,'g')

plt.xticks(ticks=range(len(countries)),labels=countries, rotation='vertical')

plt.show()
```

## Bar Plot

```python
country_cnt = df['Country'].value_counts().to_dict()

countries = list(country_cnt.keys())
cnt = list(country_cnt.values())

plt.rcParams["figure.figsize"] = (20,10)

plt.xticks(ticks=range(len(countries)),labels=countries, rotation='vertical')

plt.title("Presence of Countries")

plt.bar(countries,cnt,color=['red','blue','green','lightblue', 'brown'])

plt.show()
```

## Histogram

```python
plt.title("#DBLP Distribution")

plt.xlabel('#DBLP')

plt.rcParams["figure.figsize"] = (10,5)

plt.ylabel('Frequency')

plt.hist(df['#DBLP'],bins=100,color='green') # bins define how many ranges should be considered from min to max

plt.show()
```

## Box Plots

```python
outliers = dict(markerfacecolor = 'r', marker = 'D')
plt.title('#DBLP Distribution')
plt.boxplot(df['#DBLP'],labels=['#DBLP'],flierprops=outliers)
plt.show()
```

### Multiple blotplot in one

```python
data = df[['#DBLP','H-Index']]
outliers = dict(markerfacecolor = 'g', marker = 'o')
plt.rcParams["figure.figsize"] = (20,10)
fig,ax = plt.subplots()
plt.boxplot(data.values,labels=['#DBLP','H-index'],flierprops=outliers)
plt.show()
```

```python
data = df[['#DBLP','H-Index','Citation']]
outliers = dict(markerfacecolor = 'b', marker = 'D')
plt.rcParams["figure.figsize"] = (20,10)
fig,ax = plt.subplots()
plt.boxplot(data.values,labels=['#DBLP','H-index','Citation'],flierprops=outliers)
plt.show()
```

## Violin Plots
shows density distribution

```python
data = df[['#DBLP','H-Index']]

plt.rcParams["figure.figsize"] = (20,10)

fig,ax = plt.subplots()

plt.xticks(ticks=[1,2],labels=['#DBLP','H-Index'])

plt.violinplot(data.values)

plt.show()
```

## Scatter Plots
shows relative distribution of two variables

```python
plt.xlabel('#DBLP')
plt.ylabel('World Rank')

plt.scatter(df['#DBLP'],df['World Rank'])
```

```python
plt.xlabel('Citation')
plt.ylabel('World Rank')

plt.scatter(df['Citation'],df['World Rank'])
```

## Bubble Plots
relative distribution among 3 variables

```python
plt.xlabel('Citation')
plt.ylabel('#DBLP')

plt.scatter(df['Citation'][:200],df['#DBLP'][:200],s=df['World Rank'][:200]) # s is the size of the bubble
```
