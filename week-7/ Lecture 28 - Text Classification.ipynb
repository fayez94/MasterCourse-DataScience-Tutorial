* In this tutorial, we will see how we can train a model to classify text (here based on their sentiment)

```python
from fastai.text.all import *
```

Download data and decompressed

```python
path = untar_data(URLs.IMDB)
path.ls()
```

take train dataset

```python
(path/'train').ls()
```

 We can gather data by using the TextDataLoaders.from_folder method. The only thing we need to specify is the name of the validation folder, which is “test” (and not the default “valid”).

```python
dls = TextDataLoaders.from_folder(path, valid='test')
```

```python
dls.show_batch()
```

We can see that the library automatically processed all the texts to split then in tokens, adding some special tokens like:

* xxbos to indicate the beginning of a text
* xxmaj to indicate the next word was capitalized

Then, we can define a Learner suitable for text classification in one line:

* We use the AWD LSTM architecture
* drop_mult is a parameter that controls the magnitude of all dropouts in that model
* we use accuracy to track down how well we are doing


```python
learn = text_classifier_learner(dls, AWD_LSTM, drop_mult=0.5, metrics=accuracy)
```

We can then fine-tune our pretrained model:

```python
learn.fine_tune(4, 1e-2)
```

To see how well our model is doing, we can use the show_results method:

```python
learn.show_results()
```

And we can predict on new texts quite easily:

```python
learn.predict("I really liked that movie!")
```

```python
learn.predict("That movie was bad")
```
